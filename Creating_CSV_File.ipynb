```python
import pandas as pd
from faker import Faker
import random

fake= Faker()
departments=["HR","Finance","IT","Marketing","Sales","Operations"]
records=[]

for i in range(1,1001):
  employee_id=i
  employee_name=fake.name()
  department=random.choice(departments)
  salary=random.randint(30000,100000)
  age=random.randint(20,60)
  records.append([employee_id,employee_name,age,department,salary])

emp_df=pd.DataFrame(records,columns=["Employee_ID","Employee_Name","Age","Department","Salary"])
emp_df.show()
```

```python
import pandas as pd
from faker import Faker
import random

fake= Faker()
departments=["HR","Finance","IT","Marketing","Sales","Operations"]
records=[]

for i in range(1,1001):
  employee_id=f"E{i:05d}"
  employee_name=fake.name()
  department=random.choice(departments)
  salary=random.randint(30000,100000)
  age=random.randint(20,60)
  records.append([employee_id,employee_name,age,department,salary])

emp_df=pd.DataFrame(records,columns=["Employee_ID","Employee_Name","Age","Department","Salary"])
display(emp_df)
emp_df.to_csv("employee_data.csv",index=False)
```
